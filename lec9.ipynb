```python
import numpy as np 
import matplotlib
from tkinter import *
from tkinter import filedialog

class myclass:
    def __init__(self,window):
        self.window = window
        frame1 = Frame(window)
        button1 = Button(frame1,text="New button",command=self.buttonpress)
        self.labeltext = StringVar()
        self.labeltext.set("Unpressed")
        label1 = Label(frame1,textvariable=self.labeltext)
        label1.pack()
        button1.pack()
        frame1.pack()
        
        button2 = Button(window,text="File chooser",command=self.choosefile)
        button2.pack()
    def buttonpress(self):
        print("You pressed the button!")
        self.labeltext.set("Pressed!")
    def choosefile(self):
        tmp = filedialog.askopenfilename(initialdir='/',title="Select file!")
        print(tmp)
root = Tk()
my_gui = myclass(root)
root.mainloop()
```

```python
from matplotlib.figure import Figure
from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg
class sorter:
    def __init__(self,window,n):
        self.n = n
        self.window = window
        self.data = np.argsort(np.random.uniform(size=n))
        self.swaps = self.selection()
        
        fig = Figure(figsize=(6,6))
        fig.patch.set_facecolor('black')
        fig.subplots_adjust(left=0,bottom=0,top=1,right=1)
        ax = fig.add_subplot(111)
        ax.scatter(np.arange(n),self.data,color='yellow',edgecolors='black')
        self.ax = ax
        ax.axis('off')
        
        canvas = FigureCanvasTkAgg(fig,master=window)
        canvas.get_tk_widget().pack()
        canvas.draw()
        self.canvas = canvas
        self.update()
    def update(self):
        if not self.swaps: return
        (i,j) = self.swaps.pop(0)
        self.ax.scatter((i,j),(self.data[i],self.data[j]),color='black')
        (self.data[i],self.data[j]) = (self.data[j],self.data[i])
        self.ax.scatter((i,j),(self.data[i],self.data[j]),color='yellow',edgecolors='black')
        self.canvas.draw()
        root.after(1,self.update)
    def bubble(self):
        swaps = []
        change=True
        y = self.data.copy()
        while(change):
            change=False
            for i in range(self.n-1):
                if y[i]>y[i+1]:
                    swaps.append((i,i+1))
                    (y[i],y[i+1]) = (y[i+1],y[i])
                    change = True
        return(swaps)
    def selection(self):
        swaps = []
        y = self.data.copy()
        for i in range(self.n-1):
            argval=i
            for j in range(i+1,self.n):
                if y[j]<y[argval]:
                    argval = j
            swaps.append((i,argval))
            (y[i],y[argval]) = (y[argval],y[i])
        return(swaps)
    
root = Tk()
my_gui = sorter(root,100)
root.mainloop()
```

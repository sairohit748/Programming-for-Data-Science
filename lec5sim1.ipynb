```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
import statsmodels.formula.api as smf
```

```python
sig = 0.5
nsim = 1000
Tvec = [10,25,50,100,250]
nvec = [1,2,3,5,10]
holding_place_3b = np.zeros((nsim,len(Tvec),len(nvec)))
for ndx,n in enumerate(nvec):
    for tdx,T in enumerate(Tvec):
        for isim in range(nsim):
            y = np.random.normal(0,sig,size=(T,1))
            x = np.random.normal(0,sig,size=(T,n))
            x = sm.add_constant(x)
            model = smf.OLS(y,x).fit()
            #np.vstack((x.mean(0),x.std(0))).T
            holding_place_3b[isim,tdx,ndx] = model.rsquared
print(holding_place_3b.mean(0))
```

```python
pd.DataFrame(holding_place_3b.mean(0))
```

```python
sig = 0.5
nsim = 1000
Tvec = [10,25,50,100,250]
nvec = [1,2,3,5,10]
holding_place = np.zeros((nsim,len(Tvec),len(nvec)))
for ndx,n in enumerate(nvec):
    for tdx,T in enumerate(Tvec):
        for isim in range(nsim):
            y = np.random.normal(0,sig,size=(T,1))
            x = np.random.normal(0,sig,size=(T,n))
            for t in range(1,T):
                y[t] += y[t-1]
                x[t,:] += x[t-1,:]
            x = sm.add_constant(x)
            model = smf.OLS(y,x).fit()
            #np.vstack((x.mean(0),x.std(0))).T
            holding_place[isim,tdx,ndx] = model.rsquared
print(holding_place.mean(0))
```

```python
nsim = 500
Tvec = [10,25,50,100]
rvec = [0.3,0.4,0.5,0.6,0.7,0.8,0.9]
svec = [0.5,1,2]

outp = np.zeros((len(Tvec)*len(rvec)*len(svec),nsim,5))
rowcnt = 0
for T in Tvec:
    for sig in svec:
        for rho in rvec:
            for isim in range(nsim):
                y = np.random.normal(0,sig,size=(T+100,1))
                for t in range(1,T+100):
                    y[t] += rho*y[t-1]
                y = y[100:]
                y_0 = y[2:T]
                y_1 = y[1:T-1]
                xdata = sm.add_constant(y_1)
                outp[rowcnt,isim,0] = T
                outp[rowcnt,isim,1] = sig
                outp[rowcnt,isim,2] = rho
                outp[rowcnt,isim,3] = smf.OLS(y_0,y_1).fit().params
                outp[rowcnt,isim,4] = smf.OLS(y_0,xdata).fit().params[1]
            rowcnt += 1
outp.mean(1)
#0.5657013738527894*0.75**2+0.25
```

```python
smf.OLS(y_0,xdata).fit().params[1]
```

```python
pd.DataFrame(outp.mean(1))
```

```python
df = pd.DataFrame(np.hstack((outp.mean(1),outp.var(1)[:,3:])))
df.columns = ('sample','sig','rho','E_b_nocon','E_b_con','V_b_nocon','V_b_con')
df
```

```python
model1 = smf.ols('E_b_nocon~sample+rho+sig',data=df).fit()
model1.summary()
```

```python
model15 = smf.ols('np.log(E_b_nocon)~np.log(sample)+np.log(sample-1)+np.log(rho)',data=df).fit()
model15.summary()
model16 = smf.ols('np.log(I(rho-E_b_nocon))~np.log(sample)+np.log(rho)',data=df).fit()
model16.summary()
np.exp(0.1861)
```

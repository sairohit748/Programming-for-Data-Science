```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
import statsmodels.formula.api as smf
import math as mth
```

```python

nsim = 500
Tvec = [10,25,50,100]
rvec = [0.3,0.4,0.5,0.6,0.7,0.8,0.9]
svec = [0.5,1,2]

outp = np.zeros((len(Tvec)*len(rvec)*len(svec),nsim,8))
rowcnt = 0
for T in Tvec:
    for sig in svec:
        for rho in rvec:
            for isim in range(nsim):
                y = np.random.normal(0,sig,size=(T+100,1))
                for t in range(1,T+100):
                    y[t] += rho*y[t-1]
                y = y[100:]
                y_0 = y[2:T]
                y_1 = y[1:T-1]
                xdata = sm.add_constant(y_1)
                outp[rowcnt,isim,0] = T
                outp[rowcnt,isim,1] = sig
                outp[rowcnt,isim,2] = rho
                outp[rowcnt,isim,3] = smf.OLS(y_0,y_1).fit().params
                outp[rowcnt,isim,4] = smf.OLS(y_0,xdata).fit().params[1]
                outp[rowcnt,isim,5] = (outp[rowcnt,isim,3]*outp[rowcnt,isim,0])/(outp[rowcnt,isim,0]-1.6514)
                if((outp[rowcnt,isim,0]-(4*0.8027))**2)-(2.57730916+(4*outp[rowcnt,isim,4]*outp[rowcnt,isim,0]))<0.0:
                    outp[rowcnt,isim,6] = outp[rowcnt,isim,4]
                else:
                    outp[rowcnt,isim,6] = (-(outp[rowcnt,isim,0]-(4*0.8027))+mth.sqrt(((outp[rowcnt,isim,0]-(4*0.8027))**2)-(2.57730916+(4*outp[rowcnt,isim,4]*outp[rowcnt,isim,0]))))/(2*0.8027)
                    outp[rowcnt,isim,7] = (-(outp[rowcnt,isim,0]-(4*0.8027))-mth.sqrt(((outp[rowcnt,isim,0]-(4*0.8027))**2)-(2.57730916+(4*outp[rowcnt,isim,4]*outp[rowcnt,isim,0]))))/(2*0.8027)
            rowcnt += 1
outp.mean(1)
#0.5657013738527894*0.75**2+0.25
```
